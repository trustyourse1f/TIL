```python
import pandas as pd
import numpy as np

bike_weather = pd.read_csv('bike_weather.csv')
bike_weather
```

```python
bike_weather['over_500'] = 1
bike_weather.loc[bike_weather.Count < 500, 'over_500'] = 0
bike_weather.to_csv('bike_weather.csv', encoding = 'utf-8-sig', index = False)
```

```python
bike_weather
```

```python
from sklearn.model_selection import train_test_split

X = bike_weather[['cum_precipitation', 'humidity', 'temp', 'wind']]
y = bike_weather.over_500
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 123)
```

```python
import statsmodels.api as sm

X1 = sm.add_constant(X_train)
logit_model = sm.Logit(y_train, X1)
result = logit_model.fit()
print(result.summary())
```

Pseudo R-squ : 회귀분석이 아니므로 Pseudo라는 표현 사용, 회귀처럼 계산하면 0.4227 정도의 R제곱값이 나온다는 의미  
LLR p-value : 1.530e-05이므로 아주 작은 값 : 통계적 유의성에 문제없음  
독립변수들의 p-value : humidity를 제외한 독립변수들은 통계성 유의성이 부족함  

유의성이 확인되지 않은 독립변수  
일반적인 통계에서는 p-value의 유의성이 확인되지 않은 독립변수를 제외  
> 머신러닝에서는 시각이 다름  

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
print('Train set 정확도 : %.2f'%log_reg.score(X_train, y_train))
print('Test set 정확도 : %.2f'%log_reg.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report

y_pred = log_reg.predict(X_test)
print(classification_report(y_test, y_pred))
```
