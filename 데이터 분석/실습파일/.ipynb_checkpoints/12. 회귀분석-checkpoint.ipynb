```python
import pandas as pd

weather = pd.read_csv('weather.csv')
weather
```

```python
bike_data2 = pd.read_csv('bike_data2.csv')
bike_data2
```

weather : 시분단위  
bike_data2 : 시간단위  

두 데이터를 결합하려면 단위를 통일해야 한다.

### weather : 시간 단위로 대여건수 평균 요약

```python
import numpy as np

new_weather = pd.pivot_table(weather, index = ['date', 'time'], 
                             values = ['temp', 'cum_precipitation', 'humidity', 'insolation', 'sunshine', 'wind', 
                                       'wind_direction', 'sea_lvl_pressure', 'pressure'], aggfunc = np.mean)

new_weather
```

date가 index로 들어가 있음

```python
new_weather = new_weather.reset_index()
new_weather
```

새로운 index가 부여되었고 date가 일반 column으로 접근 가능해졌다

### bike_data2 : 일자와 시간대 별로 대여건수 요약

```python
new_bike = pd.pivot_table(bike_data2, index = ['Date_out', 'Time_out'], values = ['Distance'], aggfunc = len)
new_bike = new_bike.reset_index()
new_bike
```

> 컬럼 이름 변경

```python
new_bike.rename(columns = {'Distance' : 'Count'}, inplace = True)
new_bike
```

```python
bike_weather = pd.merge(new_bike, new_weather, left_on = ['Date_out', 'Time_out'], right_on = ['date', 'time'])
bike_weather.to_csv('bike_weather.csv', encoding = 'utf-8-sig')
bike_weather
```

### 회귀식을 찾는 함수
1. scipy 패키지의 stats모듈에 있는 함수
2. statsmodels 패키지의 회귀분석 함수

```python
from scipy import stats

stats.linregress(bike_weather.temp, bike_weather.Count)
```

계수 : 38.793750423978715, 절편 : -320.14608008876144  
선형회귀식 : Count = 38.793750423978715 * temp - 320.14608008876144  
rvalue : R squared 값으로 사용

회귀분석의 귀무가설 : 회귀관계가 없다. 회귀식의 기울기가 0이다

R제곱
- 회귀식을 판단하는 중요한 결과값
- 결정계수
- rvalue의 제곱값

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(bike_weather.temp, bike_weather.Count)
print('R-squared : %f'%r_value**2)
```

### R제곱(결정계수)
회귀식이 실제 관찰된 값을 얼마나 설명하는지를 의미  
데이터 세트에 있는 x 값을 회귀식에 넣었을 때 계산된 y값이 실제 데이터에 존재하는 y 값 대비 얼마나 되는지를 말함  
= **설명력**  
0에서 1사이의 값  
1에 가까울수록 설명력이 높아 좋은 회귀식으로 평가함  

```python
import statsmodels.api as sm

X0 = bike_weather.temp
X1 = sm.add_constant(X0)
y = bike_weather.Count
model = sm.OLS(y, X1)
result = model.fit()
print(result.summary())
```
