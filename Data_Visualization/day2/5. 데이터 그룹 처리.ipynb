## 5. 데이터 그룹 처리

```python
import numpy as np
import pandas as pd

ipl_data = {'Team' : ['Riders', 'Riders', 'Devils', 'Devils', 'Kings', 'Kings', 
                      'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
           'Rank' : [1, 2, 2, 3, 3, 4, 1, 1, 2, 4, 1, 2],
           'Year' : [2014, 2015, 2014, 2015, 2014, 2015, 2016, 2017, 2016, 2014, 2015, 2017],
           'Points' : [876, 789, 863, 673, 741, 812, 756, 788, 694, 701, 804, 690]}
df = pd.DataFrame(ipl_data)

df
```

```python
df.groupby('Team')['Points'].sum()
```

```python
c_df = df.groupby(['Team', 'Year'])['Points'].sum()
```

```python
c_df
```

```python
c_df.index
```

```python
c_df['Devils':'Kings']
```

```python
c_df.unstack()
```

```python
c_df.swaplevel().sort_index()
```

```python
c_df.sum(level=0)
```

```python
dc_df = df.groupby('Team')
dc_df.get_group('Riders')
```

```python
dc_df.agg(min)
```

```python
dc_df.agg(np.mean)
```

```python
dc_df.agg(np.max)
```

```python
f = lambda x : (x - x.mean())/x.std()
```

```python
dc_df.transform(f)
```

```python
dc_df.filter(lambda x : len(x) >= 3)
```

```python
dc_df.filter(lambda x : x['Points'].max() >= 800)
```

```python
ck = dc_df.agg(np.max)
ck
```

```python
data = {
    'Team' : ['Devils', 'Kings', 'Riders', 'Royals', 'Kings', 'A'],
    'num' : [3, 4, 6, 2, 8, 10],
    'a_id' : ['d1', 'd2', 'd3', 'd4', 'd5', 'd6']
}

ck2 = pd.DataFrame(data)
ck2
```

```python
pd.merge(left = ck, right = ck2, how = 'inner', on = 'Team')
```

```python
pd.merge(ck, ck2, on = 'Team', how = 'right')
```

```python
pd.merge(ck, ck2, on = 'Team', how = 'left')
```

```python
pd.merge(ck, ck2, on = 'Team', how = 'outer')
```

```python
ck
```
